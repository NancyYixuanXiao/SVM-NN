```python
import pandas as pd
wdbc_data = pd.read_csv('WDBC_raw.csv')
X_train = wdbc_data.drop('diagnosis', axis=1).drop('Unnamed: 32', axis=1)
X_train.head()
```

```python
import matplotlib.pyplot as plt
variance = []
from sklearn.decomposition import PCA
for k in range(3,16):
    pca = PCA(n_components=k)
    pca.fit(X_train)
    # X_train_pca = pca.transform(X_train)
    variance.append(pca.explained_variance_ratio_.mean())
plt.plot(list(range(3,16)),variance)
plt.xlabel('Number of components')
plt.ylabel('variance')
plt.show()
```

```python
pca = PCA(n_components=5)
X_train_pca = pca.fit_transform(X_train)
X_train_pca
```
